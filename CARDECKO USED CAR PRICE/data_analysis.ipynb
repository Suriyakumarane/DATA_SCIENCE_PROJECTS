#### **Univariate Analysis**

##### **Categorical Varaibles - Count**

```python
categories = ['manufacturer',
'model_year',
'body_type',
'transmission_type',
'fuel_type',
'total_owners',
'insurance_type',
'seats',
'number_of_cylinders',
'valves_per_cylinder',
'super_charger',
'turbo_charger',
'drive_type',
'steering_type',
'location'
]

for i in categories:
    print(df[i].value_counts())
    print('______________________________________________________________________')
```

```python
for category in categories:
    fig = px.bar(x = df[category].value_counts().index, y = df[category].value_counts().values, title = f'Count of {category}s')
    fig.show()
```

##### **Numerical Variables - Distribution**

```python
numericals = ['total_kms', 'price' ,'mileage(kmpl)' ,'engine(CC)', 'cargo_volume']
```

```python
for numeric in numericals:
    fig = px.histogram(x = df[numeric], title = f'Distribution of {numeric}')
    fig.show()
```

##### **Outliers in Numeric Variables**

```python
for numeric in numericals:
    fig = px.box(df[numeric], title = f'Outliers in {numeric}')
    fig.show()
```
